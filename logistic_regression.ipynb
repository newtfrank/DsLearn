# USING LOGISTIC REGRESSION

```python
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
# create data
array = np.random.randint(10, 100, (100000, 4))
df = pd.DataFrame(array, columns=['c1', 'c2', 'c3', 'c4'])

df['gender'] = np.random.choice(['male', 'female', 'transgender'], (100000, 1))
df.head()
```

# DESCRIPTIVE ANALYSIS

```python
df.describe()
```

# VISUALIZATION

```python
# SEABORN BAR PLOT

sns.barplot(data=df, y='gender', x='c4')
```

```python
# SEABORN BOXPLOT
sns.boxplot(data=df, y='gender', x='c2')
```

```python
# SEABORN LINEPLOT
sns.lineplot(data=df, x='c3', y='c4')
```

# MODEL CREATION

```python
# SPLITTING THE DATA FOR MODELING

x = ['c4', 'c2']
x_data = df[x]
y_data = df.gender

x_test, x_train, y_test, y_train = train_test_split(x_data, y_data,test_size=0.2,random_state=1)
```

```python
#INITIALIZE MODEL
model = LogisticRegression()
```

```python
model.fit(x_train, y_train)
```

```python
pred = model.predict(x_test)
```

```python
score = classification_report(y_test,pred)
print(score)
```

```python
accuracy = accuracy_score(y_test,pred)
print(accuracy)
```

```python
confusion = confusion_matrix(y_test, pred)
print(confusion)
```
